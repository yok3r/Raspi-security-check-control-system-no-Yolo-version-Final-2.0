```python
import cv2
import numpy as np
import pickle
import matplotlib.pyplot as plt
import time

#load the model
with open('cnn_model_450_3160.pkl', 'rb') as p_f:
    model = pickle.load(p_f)

cap = cv2.VideoCapture(0)
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

count_eva = 0
count_edgar = 0
while True:

    ret, frame = cap.read()
    overlay = frame.copy()
    
    faces = face_classifier.detectMultiScale(frame, 1.3, 5)
    for (x,y,w,h) in faces:
        magin = 100
        top_left_x = x-magin
        top_left_y = y-magin
        bottom_right_x = x+w+magin
        bottom_right_y = y+w+magin

        cropped = frame[top_left_y:bottom_right_y, top_left_x:bottom_right_x]
        
        
        try:            
            img = cv2.resize(cropped, (450,450))
            im2arr = np.array(img)
            im2arr1 = im2arr.reshape(1,450,450,3)

            time.sleep(0.05) 
            predictions = model.predict(im2arr1)
            predictions = predictions[0][0]
            
            alpha = 0.5
            
            text = str(predictions)
            if predictions < 0.19:
                text = "Eva"
                count_eva+=1
                
                cv2.rectangle(frame, (x,y),(x+w,y+w), (127,0,255), 2)
                cv2.rectangle(overlay, (x,y-90),(x+w,y), (127,0,255), -1)
                frame = cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0)
                
                cv2.putText(frame,text,(int(top_left_x+w/2),top_left_y+60), cv2.FONT_HERSHEY_COMPLEX, 1.5 ,(255,255,255), 2)
                precisio = (1-predictions)*100
                print(precisio)
                cv2.putText(frame,str(int(precisio))+"%",(int(top_left_x+w/2),top_left_y+90), cv2.FONT_HERSHEY_COMPLEX, 1 ,(255,255,255), 2)
                
            elif predictions > 0.9:
                text = "Edgar"
                count_edgar
                
                cv2.rectangle(frame, (x,y),(x+w,y+w), (255,0,50), 2)
                cv2.rectangle(overlay, (x,y-90),(x+w,y), (255,0,50), -1)
                frame = cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0)
                
                cv2.putText(frame,text,(int(top_left_x+w/2),top_left_y+60), cv2.FONT_HERSHEY_COMPLEX, 1.5 ,(255,255,255), 2)
                precisio = 100-(1-predictions)*100
                print(precisio)
                cv2.putText(frame,str(int(precisio))+"%",(int(top_left_x+w/2),top_left_y+90), cv2.FONT_HERSHEY_COMPLEX, 1 ,(255,255,255), 2)
            else:
                text = "Unknown"
                
                #cv2.rectangle(frame, (x,y),(x+w,y+w), (127,255,255), 2)
                cv2.rectangle(overlay, (x,y-90),(x+w,y), (127,255,255), -1)
                frame = cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0)
                
                alpha = 0.2
                cv2.rectangle(overlay, (x,y),(x+w,y+h), (127,255,255), -1)
                frame = cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0)
                cv2.rectangle(overlay, (x,y),(x+w,y+h), (127,255,255), 2)
                
                cv2.putText(frame,text,(int(top_left_x+w/2),top_left_y+60), cv2.FONT_HERSHEY_COMPLEX, 1.5 ,(255,255,255), 2)
                
            #text_pred = str(predictions)
            #cv2.putText(frame,text,(30,70), cv2.FONT_HERSHEY_COMPLEX, 1 ,(0,0,0), 3)
            #cv2.putText(frame,text_pred,(30,110), cv2.FONT_HERSHEY_COMPLEX, 1,(0,0,0), 6)
            
        except Exception as e:
            print(str(e))

        
    cv2.imshow('Object Detector using ORB', frame)
    
    if cv2.waitKey(1) == 13: #13 is the Enter Key
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
